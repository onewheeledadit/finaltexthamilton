# GenSim Topic Modeling

```python
# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(corp_list)
```

```python
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in corp_list]
```

```python
corpus
```

```python
# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=10, id2word = dictionary, passes=30)
```

```python
print(ldamodel.print_topics(num_topics=10, num_words=10))
```

# SK Learn Topic Modeler 

```python
files = glob.glob('/Users/kevinzhai/Desktop/Text Analytics/finaltexthamilton/txts/*.txt')
corp_list = []

for filename in files:
        file = open(filename, 'r')
        corp = file.read()
        corp_list.append(corp)
        file.close()
```

```python
n_samples = 2000
n_features = 1000
n_topics = 10
n_top_words = 20

def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()

tf_vectorizer = CountVectorizer(max_df=0.60, min_df=5,
                                max_features=n_features,
                                stop_words='english')
tf = tf_vectorizer.fit_transform(corp_list)

lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)

```

```python
lda.fit(tf)
```

```python
print("\nTopics in LDA model:")
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
```
